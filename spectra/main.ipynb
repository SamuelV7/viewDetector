```python
import torch
import matplotlib.pyplot as plt
```

```python
# Generate some example data
X = torch.randn(100, 50)  # 100 samples, 50 features

# Center the data
X_centered = X - X.mean(dim=0)

# Perform SVD
U, S, V = torch.svd(X_centered)

# Project onto the first k principal components
k = 50
X_pca = torch.matmul(X_centered, V[:, :k])
```

```python
X_pca_np = X_pca.numpy()

# Plot the data projected onto the first two principal components
plt.figure(figsize=(8, 6))
plt.scatter(X_pca_np[:, 0], X_pca_np[:, 1], c='blue', alpha=0.6, edgecolor='k')
plt.title("PCA Projection onto First Two Components")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.show()
```
